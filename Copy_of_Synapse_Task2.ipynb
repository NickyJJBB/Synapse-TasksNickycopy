## Task 2.2 - Operations

Write a code to give me the number of rows that have the release_year as 2021 and type as Movie.

```python
filtered_df = df[(df['release_year'] == 2021) & (df['type'] == 'Movie')] # Your output will be 146, if correct
num=len(filtered_df)
num
```

Write a code to create a list of all the movies that have the word 'Avengers' in it.

```python
filtered_movies=df[(df['type']=='Movie') & df['title'].str.contains('Avengers')]# Your output will contain 2 movies, if correct
alist=list(filtered_movies['title'])
alist
```

Display the number of Movies/TV shows per country.

```python
list1=[]
for value in df['country']:
  list1.append(value.split(', '))

oned=[]
for item in list1:
  for i in item:
    oned.append(i)

from collections import Counter
Counter(oned).most_common()
```

Display a new dataframe, that contains only those movies that are present in the Top 5 countries.

(Don't forget to reset the indices again)

```python
new_df=df[((df['type']=='Movie') | (df['type']=='TV Show')) & ((df['country']=='United States') | (df['country']=='India') | (df['country']=='United Kingdom') | (df['country']=='Canada') | (df['country']=='Spain') )]
#The length of new dataframe will be 3102, if correct
new_df.reset_index(inplace=True)
new_df
```

## Task 2.3 - Chart

Use the new dataframe to create a Pie Chart on the basis of number of Movies/TV Shows per country.

```python
import matplotlib.pyplot as plt
pie_df=new_df['country'].value_counts()
pie_df.plot(kind='pie',labels=pie_df.index,autopct='%1.1f%%')

plt.title("Distribution of Movies and TV Shows per country")
plt.show()
```
